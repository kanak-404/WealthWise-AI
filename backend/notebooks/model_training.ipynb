```python
# Step 1: Imports
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import pickle

# Step 2: Create synthetic dataset
data = []

for _ in range(1000):
    income = np.random.randint(10000, 100000)
    expense = np.random.randint(5000, income)
    savings = income - expense
    savings_rate = savings / income
    ratio = expense / income

    # Label logic (simple rule-based)
    approved = 1 if savings_rate > 0.2 and ratio < 0.8 else 0

    data.append([income, expense, savings, savings_rate, ratio, approved])

df = pd.DataFrame(data, columns=[
    "income",
    "total_spent",
    "savings",
    "savings_rate",
    "expense_to_income_ratio",
    "approved"
])

# Step 3: Train model
X = df.drop("approved", axis=1)
y = df["approved"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = DecisionTreeClassifier()
model.fit(X_train, y_train)

# Step 4: Evaluate
preds = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, preds))

# Step 5: Save model
with open("../model/model.pkl", "wb") as f:
    pickle.dump(model, f)

print("Model saved successfully!")
```
